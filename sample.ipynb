```python


# Generate x values from 0 to 2*pi (one complete cycle of the sine wave)
x = np.linspace(0, 2 * np.pi, 1000)

# Calculate the corresponding y values (sine of x)
y = np.sin(x)

# Create a plot
plt.figure(figsize=(10, 4))
plt.plot(x, y, label='Sine Wave', color='blue')

# Add labels and a legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine Wave')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
